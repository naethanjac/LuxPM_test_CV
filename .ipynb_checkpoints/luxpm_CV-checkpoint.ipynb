```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial import distance
```

```python
img_path = 'input.jpg'
```

```python
img = cv2.imread(img_path)
img.shape
```

```python
tx = 0.25*img.shape[1]
ty = 0.25*img.shape[0]
#transformation matrix
tfms_mat_tx = np.array([
    [1, 0, tx],
    [0, 1, 0]
])
tfms_mat_ty = np.array([
    [1, 0, 0],
    [0, 1, -ty]    #negative since (0,0) in opencv is top left corner instead of bottom left
])
```

```python
#assuming output image size should be same as original, clipping the image
tx_shift = cv2.warpAffine(img, tfms_mat_tx, (img.shape[1], img.shape[0]))
ty_shift = cv2.warpAffine(img, tfms_mat_ty, (img.shape[1], img.shape[0]))
```

```python
plt.imshow(tx_shift)  
```

```python
plt.imshow(ty_shift)
```

```python
centre = (img.shape[1]/2, img.shape[0]/2)
tfms_mat_rot_90 = cv2.getRotationMatrix2D(centre,angle=90,scale=1)
tfms_mat_rot_90_neg = cv2.getRotationMatrix2D(centre,angle=-90,scale=1) 
rot_90 = cv2.warpAffine(img, tfms_mat_rot_90 , (img.shape[1], img.shape[0]))
rot_90_neg = cv2.warpAffine(img, tfms_mat_rot_90_neg , (img.shape[1], img.shape[0]))
```

```python
plt.imshow(rot_90)
```

```python
plt.imshow(rot_90_neg)
```

```python
img = np.float64(img)
for i in range(img.shape[1]):
    for j in range(img.shape[0]):
        dst = distance.euclidean(centre, (i,j))
        if dst<50:
            b_pct = 1+(50 - dst)/100 #brightness percentage
            img[j,i,:] *= b_pct
img = np.uint8(np.clip(img,0,255))
```

```python
plt.imshow(img)
```

```python
cv2.imwrite("1.png",tx_shift)
cv2.imwrite("2.png",ty_shift)
cv2.imwrite("3.png",rot_90)
cv2.imwrite("4.png",rot_90_neg)
cv2.imwrite("5.png",img)
```
